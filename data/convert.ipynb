```python
import pandas as pd
```

```python
pisa_historic = pd.read_csv("OECD PISA data.csv")

pisa_historic = pisa_historic[pisa_historic['SUBJECT'] == 'TOT']

pisa_historic = pisa_historic.groupby(['LOCATION', 'TIME'])['Value'].mean().reset_index()

pisa_historic = pisa_historic.rename(columns={
    'LOCATION': 'geo', 
    'TIME': 'year', 
    'Value': 'pisa_score'
})
```

```python
# Dicionário de conversão (ISO Alpha-2 para ISO Alpha-3)
iso2_to_iso3 = {
    'SG': 'SGP', 'CN': 'CHN', 'JP': 'JPN', 'TW': 'TWN', 'KR': 'KOR', 'HK': 'HKG',
    'EE': 'EST', 'CA': 'CAN', 'IE': 'IRL', 'CH': 'CHE', 'AU': 'AUS', 'FI': 'FIN',
    'NZ': 'NZL', 'GB': 'GBR', 'PL': 'POL', 'CZ': 'CZE', 'DK': 'DNK', 'US': 'USA',
    'SE': 'SWE', 'BE': 'BEL', 'AT': 'AUT', 'SI': 'SVN', 'LV': 'LVA', 'DE': 'DEU',
    'NL': 'NLD', 'FR': 'FRA', 'PT': 'PRT', 'ES': 'ESP', 'HU': 'HUN', 'LT': 'LTU',
    'IT': 'ITA', 'NO': 'NOR', 'HR': 'HRV', 'VN': 'VNM', 'IL': 'ISR', 'TR': 'TUR',
    'MT': 'MLT', 'SK': 'SVK', 'IS': 'ISL', 'RS': 'SRB', 'UA': 'UKR', 'BN': 'BRN',
    'GR': 'GRC', 'CL': 'CHL', 'RO': 'ROU', 'AE': 'ARE', 'UY': 'URY', 'QA': 'QAT',
    'BG': 'BGR', 'MD': 'MDA', 'KZ': 'KAZ', 'MX': 'MEX', 'MN': 'MNG', 'ME': 'MNE',
    'MY': 'MYS', 'CR': 'CRI', 'CY': 'CYP', 'PE': 'PER', 'CO': 'COL', 'BR': 'BRA',
    'JM': 'JAM', 'AR': 'ARG', 'TH': 'THA', 'SA': 'SAU', 'GE': 'GEO', 'AZ': 'AZE',
    'PA': 'PAN', 'MK': 'MKD', 'ID': 'IDN', 'AL': 'ALB', 'GT': 'GTM', 'PS': 'PSE',
    'SV': 'SLV', 'PY': 'PRY', 'JO': 'JOR', 'MA': 'MAR', 'PH': 'PHL', 'UZ': 'UZB',
    'DO': 'DOM', 'KH': 'KHM'
}
```

```python
pisa_2022 = pd.read_csv("pisa-scores-by-country-2026.csv")

pisa_2022['geo'] = pisa_2022['flagCode'].map(iso2_to_iso3)

pisa_2022['year'] = 2022

cols_notas = ['PISAScoresMathScore_2022', 'PISAScoresScienceScore_2022', 'PISAScoresReadingScore_2022']
pisa_2022['pisa_score'] = pisa_2022[cols_notas].mean(axis=1)

pisa_2022_clean = pisa_2022[['geo', 'year', 'pisa_score']].dropna(subset=['geo'])
```

```python
pisa_master = pd.concat([pisa_historic, pisa_2022_clean], ignore_index=True)

pisa_master = pisa_master.sort_values(by=['geo', 'year'])

pisa_master.to_csv('pisa_master_dataset.csv', index=False)

pisa_master.head()
```
